```python
import pandas as pd
import re
```

```python
sexism_data = pd.read_csv('twitter_sexism_parsed_dataset.csv')
```

```python
sexism_data.head()
```

```python
sexism_data_1 = sexism_data[sexism_data['oh_label'] == 1]  # Records where oh_label is 1
sexism_data_0 = sexism_data[sexism_data['oh_label'] == 0]  # Records where oh_label is 0
```

```python
def format_id(value):
    try:
        value_int = int(float(value))  # Ensure it's converted properly
        value_str = str(value_int)  # Convert to string for formatting
        return f"{value_str[:4]}-{value_str[4:8]}-{value_str[8:12]}-{value_str[12:16]}-{value_str[16:]}"
    except ValueError:
        return None  # Handle any conversion errors gracefully
```

```python
sexism_data_1 = pd.DataFrame(sexism_data_1)
```

```python
# Apply formatting to 'index' and 'id' columns
sexism_data_1['formatted_index'] = sexism_data_1['index'].apply(format_id)
sexism_data_1['formatted_id'] = sexism_data_1['id'].apply(format_id)
```

```python
# Display the updated dataframe
sexism_data_1[['formatted_index', 'formatted_id']].head()
```

```python
def extract_and_remove_mentions_hashtags_links(text):
    usernames = re.findall(r'@\w+', text)  # Extract usernames
    hashtags = re.findall(r'#\w+', text)  # Extract hashtags
    links = re.findall(r'http[s]?://\S+', text)  # Extract links starting with http
    clean_text = re.sub(r'@\w+|#\w+|http[s]?://\S+', '', text).strip()  # Remove usernames, hashtags, and links
    return clean_text, ' '.join(usernames), ' '.join(hashtags), ' '.join(links)  # Return cleaned text and extracted words
```

```python
sexism_data_1[['cleaned_text', 'users', 'hashtags','links']] = sexism_data_1['Text'].apply(lambda x: pd.Series(extract_and_remove_mentions_hashtags_links(x)))
```

```python
# Reorder and drop unnecessary columns
sexism_data_1 = sexism_data_1[['formatted_index', 'formatted_id', 'cleaned_text', 'Annotation', 'oh_label', 'users', 'hashtags', 'links']]
```
